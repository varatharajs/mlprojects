```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Load dataset
df = pd.read_csv(r'C:\Users\SVaratharajKannan(36\Downloads\Extended_E_O_Inventory_Dataset.csv')

# Encode categorical variables
le = LabelEncoder()
df['Lifecycle_Phase'] = le.fit_transform(df['Lifecycle_Phase'])
df['Subdivision'] = le.fit_transform(df['Subdivision'])

# Define numeric columns and apply median imputation explicitly
numeric_cols = df.select_dtypes(include=np.number).columns
imputer = SimpleImputer(strategy='median')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])

# Verify there are no remaining NaN values
if df.isna().sum().any():
    print("Warning: There are still NaN values in the data.")
else:
    print("No missing values detected. Proceeding with modeling.")

# Feature Engineering
df['Month'] = pd.to_datetime(df['Date'], errors='coerce').dt.month
df['Demand_3M_Rolling_Avg'] = df.groupby('Subdivision')['Actual_Demand'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())
df['Lagged_Demand'] = df.groupby('Subdivision')['Actual_Demand'].shift(1)  # 1-month lagged demand

# Drop any rows with NaN values created by shifting or rolling calculations
df.dropna(inplace=True)

# Define features and target for regression
X_reg = df[['Forecast_Accuracy', 'Demand_Fluctuation', 'Lead_Time_Days', 'Remaining_Shelf_Life_Months', 
            'Lifecycle_Phase', 'Subdivision', 'Carrying_Cost', 'Unit_Cost', 'Demand_3M_Rolling_Avg', 'Lagged_Demand']]
y_reg = df['Excess_Stock']

# Split data
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Initialize regression models with parameter grids
param_grid_rf = {'n_estimators': [50, 100], 'max_depth': [None, 10, 20]}
rf_reg = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=3, scoring='r2')
gb_reg = GradientBoostingRegressor(random_state=42)
linear_reg = LinearRegression()

# Train and evaluate regression models
reg_results = {}
for name, model in zip(['Random Forest Regressor', 'Gradient Boosting Regressor', 'Linear Regression'], [rf_reg, gb_reg, linear_reg]):
    # Fit the model and predict
    model.fit(X_train_reg, y_train_reg)
    y_pred = model.predict(X_test_reg)
    
    # Calculate metrics
    r2 = r2_score(y_test_reg, y_pred)
    mae = mean_absolute_error(y_test_reg, y_pred)
    reg_results[name] = {'R-squared': r2, 'MAE': mae}

print("\n--- Regression Results ---")
for model_name, metrics in reg_results.items():
    print(f"Model: {model_name}")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value}")
    print("\n" + "-"*30 + "\n")
```

# Data Classification & Regression

### Step 1: Data Loading and Preprocessing

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, RandomForestClassifier
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import mean_absolute_error, r2_score, accuracy_score, f1_score, jaccard_score, confusion_matrix
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r'C:\Users\SVaratharajKannan(36\Downloads\Extended_E_O_Inventory_Dataset.csv')

# Encode categorical variables and save mappings
le_lifecycle = LabelEncoder()
df['Lifecycle_Phase'] = le_lifecycle.fit_transform(df['Lifecycle_Phase'])
lifecycle_mapping = dict(zip(le_lifecycle.classes_, le_lifecycle.transform(le_lifecycle.classes_)))

le_subdivision = LabelEncoder()
df['Subdivision'] = le_subdivision.fit_transform(df['Subdivision'])
subdivision_mapping = dict(zip(le_subdivision.classes_, le_subdivision.transform(le_subdivision.classes_)))

# Display mappings for verification
print("Lifecycle Phase Mapping:", lifecycle_mapping)
print("Subdivision Mapping:", subdivision_mapping)

# Impute missing values for numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns
imputer = SimpleImputer(strategy='median')
df[numeric_cols] = imputer.fit_transform(df[numeric_cols])

# Check for remaining NaN values
if df.isna().sum().any():
    print("Warning: NaN values are still present.")
else:
    print("No missing values detected. Proceeding with feature engineering.")
```

### Step 2: Feature Engineering

```python
# Incorporate rolling averages, lagged demand, and additional trend-related features for better seasonality capture.
# Feature Engineering: Time-Based Variables
df['Month'] = pd.to_datetime(df['Date'], errors='coerce').dt.month
df['Demand_3M_Rolling_Avg'] = df.groupby('Subdivision')['Actual_Demand'].transform(lambda x: x.rolling(window=3, min_periods=1).mean())
df['Lagged_Demand'] = df.groupby('Subdivision')['Actual_Demand'].shift(1)

# Drop rows with NaNs created by lagging and rolling
df.dropna(inplace=True)
```

### Step 3: Define Features and Targets for Regression

```python
# Define features and target for regression
X_reg = df[['Forecast_Accuracy', 'Demand_Fluctuation', 'Lead_Time_Days', 'Remaining_Shelf_Life_Months', 
            'Lifecycle_Phase', 'Subdivision', 'Carrying_Cost', 'Unit_Cost', 'Demand_3M_Rolling_Avg', 'Lagged_Demand']]
y_reg = df['Excess_Stock']
```

### Step 4: Advanced Hyperparameter Tuning for Regression Models

```python
# Split data for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

# Parameter grids for tuning
param_grid_rf = {'n_estimators': [50, 100, 150], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10]}
param_grid_gb = {'n_estimators': [50, 100, 150], 'learning_rate': [0.01, 0.1, 0.2], 'max_depth': [3, 5, 7]}

# Initialize models with grid search
rf_reg = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=3, scoring='r2')
gb_reg = GridSearchCV(GradientBoostingRegressor(random_state=42), param_grid_gb, cv=3, scoring='r2')
linear_reg = LinearRegression()

# Train and evaluate regression models
reg_results = {}
for name, model in zip(['Random Forest Regressor', 'Gradient Boosting Regressor', 'Linear Regression'], [rf_reg, gb_reg, linear_reg]):
    model.fit(X_train_reg, y_train_reg)
    y_pred = model.predict(X_test_reg)
    r2 = r2_score(y_test_reg, y_pred)
    mae = mean_absolute_error(y_test_reg, y_pred)
    reg_results[name] = {'R-squared': r2, 'MAE': mae}

print("\n--- Regression Results ---")
for model_name, metrics in reg_results.items():
    print(f"Model: {model_name}")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value}")
    print("\n" + "-"*30 + "\n")
```

### Step 5: Define and Train Classification Models

```python
# Define target for classification
y_class = (df['Excess_Stock'] > df['Excess_Stock'].median()).astype(int)

# Split data for classification
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_reg, y_class, test_size=0.2, random_state=42)

# Re-apply the original mappings to decode values back to the original form
df['Lifecycle_Phase'] = df['Lifecycle_Phase'].map({v: k for k, v in lifecycle_mapping.items()})
df['Subdivision'] = df['Subdivision'].map({v: k for k, v in subdivision_mapping.items()})

# Initialize classification models
rf_class = RandomForestClassifier(n_estimators=100, random_state=42)
log_reg = LogisticRegression(max_iter=1000)

class_results = {}
for name, model in zip(['Random Forest Classifier', 'Logistic Regression'], [rf_class, log_reg]):
    model.fit(X_train_class, y_train_class)
    y_pred_class = model.predict(X_test_class)
    accuracy = accuracy_score(y_test_class, y_pred_class)
    f1 = f1_score(y_test_class, y_pred_class)
    jaccard = jaccard_score(y_test_class, y_pred_class)
    conf_matrix = confusion_matrix(y_test_class, y_pred_class)
    class_results[name] = {
        'Accuracy': accuracy, 
        'F1 Score': f1, 
        'Jaccard Score': jaccard, 
        'Confusion Matrix': conf_matrix
    }

print("\n--- Classification Results ---")
for model_name, metrics in class_results.items():
    print(f"Model: {model_name}")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value}")
    print("\n" + "-"*30 + "\n")
```

### Step 6: Visualize Results Comparison

```python
# Visualization for Regression
reg_metrics = pd.DataFrame(reg_results).T[['R-squared', 'MAE']]
reg_metrics.plot(kind='bar', figsize=(10, 6))
plt.title("Regression Model Performance Comparison")
plt.xlabel("Model")
plt.ylabel("Performance Metric")
plt.xticks(rotation=0)
for i, (r2, mae) in enumerate(zip(reg_metrics['R-squared'], reg_metrics['MAE'])):
    plt.text(i - 0.15, r2 + 0.01, f"{r2:.2f}", color='blue')
    plt.text(i + 0.05, mae + 0.01, f"{mae:.2f}", color='orange')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()

# Visualization for Classification
class_metrics = pd.DataFrame(class_results).T[['Accuracy', 'F1 Score', 'Jaccard Score']]
class_metrics.plot(kind='bar', figsize=(10, 6))
plt.title("Classification Model Performance Comparison")
plt.xlabel("Model")
plt.ylabel("Performance Metric")
plt.xticks(rotation=0)
for i, (acc, f1, jaccard) in enumerate(zip(class_metrics['Accuracy'], class_metrics['F1 Score'], class_metrics['Jaccard Score'])):
    plt.text(i - 0.2, acc + 0.01, f"{acc:.2f}", color='blue')
    plt.text(i - 0.1, f1 + 0.01, f"{f1:.2f}", color='green')
    plt.text(i + 0.1, jaccard + 0.01, f"{jaccard:.2f}", color='red')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```
